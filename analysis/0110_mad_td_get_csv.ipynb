```python
from scale_rl.envs.dmc import DMC_HARD

def replace_underbar_to_hypen(env_name_list):
    for idx in range(len(env_name_list)):
        env_name_list[idx] = env_name_list[idx].replace('_', '-')
    return env_name_list
DMC_HARD = replace_underbar_to_hypen(DMC_HARD)
```

## Note from authors
Note that there is one weirdness: because the TD-MPC2 code implements frameskip with an environment wrapper, all of the step counters in the files need to be multiplied by 2x.

```python
def iter_loop(row, env_name):
    
    news = []
    for i in range(len(row.keys()) - 1):
        new_row = dict()
        new_row['value'] = row[f'reward/{i}']        
        new_row['seed'] = i
        new_row['metric'] = 'avg_return'
        new_row['exp_name'] = 'MAD-TD_UTD=8_Model=0.05'
        new_row['env_name'] = env_name
        new_row['env_step'] = int(row['_step'] * 2)
        news.append(new_row)

    return news
```

```python
env_name = 'dog-trot'
```

```python
env_to_df = {}
for env_name in DMC_HARD:
    print(f'{env_name} - num_seeds: {len(row.keys()) - 1}')
    new_df = pd.DataFrame(columns=['exp_name', 'env_name', 'seed', 'metric', 'env_step', 'value'])
    raw_df = pd.read_csv(f'../results/raw/mad-td/{env_name}.csv')

    for _, row in raw_df.iterrows():
        
        new_df = pd.concat([
            new_df, pd.DataFrame(iter_loop(row, env_name))
        ], ignore_index=True)

    env_to_df[env_name] = new_df
```

```python
eval_df = pd.DataFrame(columns=['exp_name', 'env_name', 'seed', 'metric', 'env_step', 'value'])

for v_df in env_to_df.values():
    eval_df = pd.concat([eval_df, v_df], ignore_index=True)
```

```python
DMC_STEPS = 1e6 # 1M
```

```python
overall_df = [] 

for env_name in DMC_HARD:
    env_df = ten_seeds_eval_df[ten_seeds_eval_df["env_name"] == env_name]
    if len(env_df) == 0:
        continue
    # assert max(env_df["env_step"]) == HB_STEPS
    env_df = env_df[env_df["env_step"] == DMC_STEPS]
    env_df["value"] = env_df["value"] / 1000.0
    num_seeds = env_df["seed"].nunique()

    grouped_data = env_df.groupby("env_step")["value"]

    env_steps = grouped_data.mean().index.values
    mean = float(grouped_data.mean().values)
    std_err = float(grouped_data.sem().values)

    print(f"{env_name} - number of seeds: {num_seeds}")
    print(f"{round(mean, 3)} ± {round(1.96 * std_err, 3)}")
    
    overall_df.append(env_df)
    
overall_df = pd.concat(overall_df)
mean = overall_df["value"].mean()
std_err = overall_df["value"].sem()
print(f"Overall: {round(mean, 3)} ± {round(1.96 * std_err, 3)}")
```
