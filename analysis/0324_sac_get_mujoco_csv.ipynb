```python
import os
import pandas as pd
```

```python
MUJOCO_ALL = [
    "HalfCheetah-v4",
    "Hopper-v4",
    "Walker2d-v4",
    "Ant-v4",
    "Humanoid-v4",
]
```

```python
concat_list = []
data_path = '/home/nas4_user/youngdolee/SimbaV2/results/raw/sac_mujoco/SAC_'
for env_name in MUJOCO_ALL:
    data_df = pd.read_csv(data_path + env_name + '_return.csv')
    
    eval_columns = [col for col in data_df.columns if col.startswith('eval/return/')]
    assert len(eval_columns) == 10
    data_df_melted = data_df.melt(id_vars=['step'], value_vars=eval_columns, var_name='seed', value_name='value')
    data_df_melted['seed'] = data_df_melted['seed'].apply(lambda x: eval_columns.index(x))
    # exp_name,env_name,seed,metric,env_step,value
    data_df_melted['exp_name'] = 'SAC'
    data_df_melted['env_name'] = env_name
    data_df_melted['metric'] = 'avg_return'
    data_df_melted.rename(columns = {'step' : 'env_step'}, inplace = True)
    data_df_melted = data_df_melted[['exp_name', 'env_name', 'seed', 'metric', 'env_step', 'value']]
    concat_list.append(data_df_melted)
    
sac_eval_df = pd.concat(concat_list, ignore_index=True)
sac_eval_df.to_csv('/home/nas4_user/youngdolee/SimbaV2/results/raw/sac_mujoco/sac_mujoco.csv')
```

```python
sac_mujoco_eval_df = pd.read_csv(
    "/home/nas4_user/youngdolee/SimbaV2/results/baseline/sac_mujoco.csv",
    index_col=0
)
sac_mujoco_eval_df['exp_name'] = 'sac'

sac_others_eval_df = pd.read_csv(
    "/home/nas4_user/youngdolee/SimbaV2/results/baseline/sac.csv",
    index_col=0
)
```
